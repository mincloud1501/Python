```python
# 기본 그래프 그리기
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,2,3,4])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.show()
```

```python
# 복수개의 시리즈 그리기
plt.plot([1,2,3,4],[1,2,3,4],'r-', [1,2,3,4],[3,4,5,6],'v-')
plt.show()
```

```python
# 한글 처리 (font_location 정의)
#‘matplotlib’는 폰트 관리를 위한 ‘fontmanager’와 리소스 관리를 위한 ‘rc’를 제공
from matplotlib import font_manager, rc
import matplotlib
font_location = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
plt.plot([1,2,3,4])
plt.xlabel('x축 한글표시')
plt.show()
```

```python
# 한 화면에 여러 개 그래프 그리기
# 한 화면에 여러 개의 그래프를 넣기 위해서는 `subplot(m , n, idx)’함수를 사용
plt.figure()
plt.subplot(2, 1, 1)
plt.plot([1,2,3,4], [1,2,3,4])
plt.subplot(2,1,2)
plt.plot([5,6,7,8],[5,6,7,8])
plt.show()

# 하나의 행에 여러 개의 열(column)을 표시하는 형태의 경우
plt.figure()
plt.subplot(1, 2, 1)
plt.plot([1,2,3,4], [1,2,3,4])
plt.subplot(1, 2, 2)
plt.plot([5,6,7,8],[5,6,7,8])
plt.show()
```

```python
# 그래프에 문자 삽입
plt.plot([1,2,3,4], [1,2,3,4])
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('matplotlib 활용')
plt.text(3.5, 3.0, '평균:2.5')
plt.grid(True)
plt.show()
```
